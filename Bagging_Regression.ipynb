```python
import numpy as  np
import pandas as pd
from sklearn.datasets import fetch_california_housing
```

```python
house = fetch_california_housing()
X,y = house.data,house.target
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split,GridSearchCV
```

```python
x_train,x_test,y_train,y_test =train_test_split(X,y,test_size = 0.2,random_state = 2)
```

```python
lr = LinearRegression()
dt = DecisionTreeRegressor()
knn = KNeighborsRegressor()
```

```python
lr.fit(x_train,y_train)
dt.fit(x_train,y_train)
knn.fit(x_train,y_train)
```

```python
y_pred1 = lr.predict(x_test)
y_pred2 = dt.predict(x_test)
y_pred3 = knn.predict(x_test)
```

```python
print('r2_score of lr',r2_score(y_test,y_pred1))
print('r2_score of dt',r2_score(y_test,y_pred2))
print('r2_score of knn',r2_score(y_test,y_pred3))
```

```python
from sklearn.ensemble import BaggingRegressor
bag = BaggingRegressor(random_state = 1)
bag.fit(x_train,y_train)
y_pred=bag.predict(x_test)
r2_score(y_test,y_pred)
```

```python
params = { 'estimator':[None,LinearRegression(),DecisionTreeRegressor(),KNeighborsRegressor()],
          'n_estimators': [50,100],
          'max_samples': [0.5,1.0],
          'max_features': [0.5,1.0],
          'bootstrap': [True, False],
          'bootstrap_features': [True, False]
           }
```

```python
grid = GridSearchCV(BaggingRegressor(random_state=1,n_jobs=-1),param_grid = params,n_jobs = -1,cv = 5,verbose = 1)
grid.fit(x_train,y_train)
grid.best_params
```
